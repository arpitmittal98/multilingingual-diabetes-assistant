```python
def get_user_profile(user_id: str) -> dict:
    """
    Fetch the stored profile. Returns {} if the user_id isn’t found.
    Column order matches the CREATE TABLE statement.
    """
    cursor.execute("SELECT * FROM users WHERE user_id = ?", (user_id,))
    row = cursor.fetchone()
    if not row:
        return {}

    # row indices:
    # 0-user_id | 1-name | 2-age | 3-gender | 4-diabetes_type | 5-insulin_use
    # 6-family_history | 7-weight_category | 8-physical_activity
    return {
        "name"             : row[1],
        "age"              : row[2],
        "gender"           : row[3],
        "diabetes_type"    : row[4],
        "insulin_use"      : bool(row[5]),
        "family_history"   : bool(row[6]),
        "weight_category"  : row[7],
        "physical_activity": row[8]
    }
```

## RAG

```python
import faiss
import json
import numpy as np
from langdetect import detect
from sentence_transformers import SentenceTransformer
import re

# Load multilingual embedding model
encoder = SentenceTransformer("sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")

def load_index_and_passages(lang):
    if lang == "es":
        index = faiss.read_index("rag_es.index")
        with open("rag_passages_es.json", "r", encoding="utf-8") as f:
            passages = json.load(f)
    else:
        index = faiss.read_index("rag_en.index")
        with open("rag_passages_en.json", "r", encoding="utf-8") as f:
            passages = json.load(f)
    return index, passages
```

```python
def build_llama3_prompt(query, context, profile: dict, lang: str):
    # Format profile nicely
    profile_text = (
        f"Name: {profile.get('name','N/A')}, Age: {profile.get('age','N/A')}, "
        f"Gender: {profile.get('gender','N/A')}, Diabetes Type: {profile.get('diabetes_type','N/A')}, "
        f"Insulin Use: {profile.get('insulin_use','N/A')}, Family History: {profile.get('family_history','N/A')}, "
        f"Weight: {profile.get('weight_category','N/A')}, Activity: {profile.get('physical_activity','N/A')}"
    )

    return (
        "<|begin_of_text|>"
        "<|start_header_id|>system<|end_header_id|>\n"
        "You are a multilingual diabetes assistant. Respond in a kind, professional, medically respectful tone, "
        "always matching the user's language.\n"
        "Answer in 2-4 sentences. If the user already meets or misses certain risk factors, state that plainly and do not list unrelated risk factors."
        "<|start_header_id|>user<|end_header_id|>\n"
        f"USER PROFILE:\n{profile_text}\n\n"
        f"MEDICAL CONTEXT:\n{context}\n\n"
        f"Question: {query}\n"
        "<|start_header_id|>assistant<|end_header_id|>\n"
    )


def answer_query_llama3(query, user_id, top_k=5, max_context_tokens=450):
    lang = detect(query)
    # print(f"Detected language: {lang}")

    # Load FAISS index + passages
    index, passages = load_index_and_passages(lang)

    # Embed query and search
    query_embedding = encoder.encode([query]).astype("float32")
    D, I = index.search(np.array(query_embedding), top_k)
    retrieved_chunks = [passages[i]["text"] for i in I[0]]

    # Token-limited concatenation
    context = ""
    token_count = 0
    for chunk in retrieved_chunks:
        tokens = tokenizer.tokenize(chunk)
        if token_count + len(tokens) > max_context_tokens:
            break
        context += chunk + "\n\n"
        token_count += len(tokens)

    # Get profile
    profile = get_user_profile(user_id)

    # Build Llama-3 prompt
    prompt = build_llama3_prompt(query, context.strip(), profile, lang)
    # prompt = build_llama3_prompt(query, context, lang)
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    # Generate
    pad_id = tokenizer.eos_token_id
    output = model.generate(
    **inputs,
    max_new_tokens=256,
    do_sample=True,
    temperature=0.3,
    top_p=0.9,
    pad_token_id=pad_id,      # add this
    eos_token_id=tokenizer.eos_token_id
    )

    answer = tokenizer.decode(output[0], skip_special_tokens=True).split("assistant")[-1].strip()
    answer = re.sub(r"^[,\. ]+", "", answer)
    return answer
```

```python
import gradio as gr
import sqlite3
import traceback

# --- Global variables ---
user_id = None
chat_history = []
chat_session_log = ""
selected_lang = "en"

# --- SQLite setup ---
conn = sqlite3.connect("personalized_rag.db", check_same_thread=False)
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS users (
    user_id TEXT PRIMARY KEY,
    name TEXT,
    age INTEGER,
    gender TEXT,
    diabetes_type TEXT,
    insulin_use BOOLEAN,
    family_history BOOLEAN,
    weight_category TEXT,
    physical_activity TEXT,
    language TEXT
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS user_history (
    user_id TEXT,
    timestamp DATETIME DEFAULT CURRENT_TIMESTAMP,
    conversation TEXT,
    FOREIGN KEY(user_id) REFERENCES users(user_id)
)
""")
conn.commit()

# --- Translations ---
labels = {
    "en": {
        "name": "Name", "age": "Age", "gender": "Gender", "gender_opts": ["male", "female", "other"],
        "dtype": "Diabetes Type", "dtype_opts": ["none", "Type 1", "Type 2", "Pre-diabetes"],
        "insulin": "Do you use insulin?", "family": "Family history of diabetes?",
        "weight": "Weight category", "weight_opts": ["underweight", "normal", "overweight", "obese"],
        "activity": "Physical activity level", "activity_opts": ["active", "moderate", "inactive"],
        "yes": "Yes", "no": "No", "submit": "Submit", "chat_label": "Ask your question",
        "status": "✅ Thanks! You can now chat below. Type 'Exit' to end the chat."
    },
    "es": {
        "name": "Nombre", "age": "Edad", "gender": "Género", "gender_opts": ["hombre", "mujer", "otro"],
        "dtype": "Tipo de diabetes", "dtype_opts": ["ninguno", "Tipo 1", "Tipo 2", "Prediabetes"],
        "insulin": "¿Usas insulina?", "family": "¿Antecedentes familiares de diabetes?",
        "weight": "Categoría de peso", "weight_opts": ["bajo peso", "normal", "sobrepeso", "obeso"],
        "activity": "Nivel de actividad física", "activity_opts": ["activo", "moderado", "inactivo"],
        "yes": "Sí", "no": "No", "submit": "Enviar", "chat_label": "Haz tu pregunta",
        "status": "✅ ¡Gracias! Ahora puedes usar el chat. Escribe 'exit' para terminar."
    }
}

# --- Save Profile ---
def save_profile(name, age, gender, diabetes_type, insulin_use, family_history, weight_category, physical_activity, lang):
    global user_id
    user_id = name.lower().replace(" ", "_")
    cursor.execute("""
        INSERT OR REPLACE INTO users (
            user_id, name, age, gender, diabetes_type, insulin_use,
            family_history, weight_category, physical_activity, language
        ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    """, (
        user_id, name, int(age), gender, diabetes_type, insulin_use == labels[lang]["yes"],
        family_history == labels[lang]["yes"], weight_category, physical_activity, lang
    ))
    conn.commit()
    return labels[lang]["status"]

# --- Chat Logic ---
def chat(query):
    global chat_history, user_id, chat_session_log
    try:
        if not user_id:
            return [("System", "⚠️ Please fill in the profile first.")]

        if query.strip().lower() == "exit":
            cursor.execute(
                "INSERT INTO user_history (user_id, conversation) VALUES (?, ?)",
                (user_id, chat_session_log.strip())
            )
            conn.commit()
            chat_history = []
            chat_session_log = ""
            user_id = None
            return [("System", "✅ Chat ended. Thank you! You may close the window or restart to begin again.")]

        # Normal turn
        answer = answer_query_llama3(query, user_id)
        chat_history.append((query, answer))
        chat_session_log += f"User: {query}\nAssistant: {answer}\n"
        return chat_history

    except Exception:
        err_detail = traceback.format_exc()
        print("❌ Full traceback:\n", err_detail)
        return chat_history + [("System", f"❌ Error:\n{err_detail[:500]}...")]

# --- UI Logic ---
with gr.Blocks() as demo:
    lang_dropdown = gr.Radio(["en", "es"], label="🌐 Choose Language / Elige idioma", value="en")

    name = gr.Text()
    age = gr.Number()
    gender = gr.Radio()
    dtype = gr.Dropdown()
    insulin = gr.Radio()
    family = gr.Radio()
    weight = gr.Dropdown()
    activity = gr.Dropdown()
    submit_btn = gr.Button()
    status_text = gr.Textbox(interactive=False, label="")

    query_input = gr.Textbox(label="Ask your question")
    chatbot = gr.Chatbot()

    def update_labels(lang):
        return {
            name: gr.update(label=labels[lang]["name"]),
            age: gr.update(label=labels[lang]["age"]),
            gender: gr.update(label=labels[lang]["gender"], choices=labels[lang]["gender_opts"]),
            dtype: gr.update(label=labels[lang]["dtype"], choices=labels[lang]["dtype_opts"]),
            insulin: gr.update(label=labels[lang]["insulin"], choices=[labels[lang]["yes"], labels[lang]["no"]]),
            family: gr.update(label=labels[lang]["family"], choices=[labels[lang]["yes"], labels[lang]["no"]]),
            weight: gr.update(label=labels[lang]["weight"], choices=labels[lang]["weight_opts"]),
            activity: gr.update(label=labels[lang]["activity"], choices=labels[lang]["activity_opts"]),
            submit_btn: gr.update(value=labels[lang]["submit"]),
            query_input: gr.update(label=labels[lang]["chat_label"])
        }

    lang_dropdown.change(update_labels, inputs=lang_dropdown, outputs=[
        name, age, gender, dtype, insulin, family, weight, activity, submit_btn, query_input
    ])

    submit_btn.click(
        fn=save_profile,
        inputs=[name, age, gender, dtype, insulin, family, weight, activity, lang_dropdown],
        outputs=status_text
    )

    query_input.submit(fn=chat, inputs=query_input, outputs=chatbot)

demo.launch(share=True)
```

```python
# Generate responses
generated_results = []

for i, q in enumerate(questions):
    print(f"Generating Q{i+1}: {q}")
    answer = answer_query_llama3(q)
    generated_results.append({"query": q, "generated_answer": answer})
```

```python
from langdetect import detect
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from rouge_score import rouge_scorer
from evaluate import load
import bert_score

# Load metrics
smooth = SmoothingFunction().method1
rouge = rouge_scorer.RougeScorer(["rouge1", "rougeL"], use_stemmer=True)
meteor = load("meteor")

# Clean helper
def clean_text(text):
    return text.strip().lower().replace('\n', ' ')

# Placeholder for metrics
all_scores = {
    "bleu": [],
    "rouge1": [],
    "rougeL": [],
    "meteor": [],
    "bertscore_f1": [],
    "langs": []
}

# Loop
for i, r in enumerate(generated_results):
    ref = clean_text(references[i])
    pred = clean_text(r["generated_answer"])

    lang = detect(ref)
    bleu = sentence_bleu([ref.split()], pred.split(), smoothing_function=smooth)
    rouge_scores = rouge.score(ref, pred)
    meteor_score = meteor.compute(predictions=[pred], references=[ref])["meteor"]
    P, R, F1 = bert_score.score([r["generated_answer"]], [references[i]], lang=lang, verbose=False)

    r["metrics"] = {
        "language": lang,
        "bleu": round(bleu, 4),
        "rouge1": round(rouge_scores["rouge1"].fmeasure, 4),
        "rougeL": round(rouge_scores["rougeL"].fmeasure, 4),
        "meteor": round(meteor_score, 4),
        "bertscore_f1": round(F1[0].item(), 4)
    }

    # Track
    all_scores["bleu"].append(bleu)
    all_scores["rouge1"].append(rouge_scores["rouge1"].fmeasure)
    all_scores["rougeL"].append(rouge_scores["rougeL"].fmeasure)
    all_scores["meteor"].append(meteor_score)
    all_scores["bertscore_f1"].append(F1[0].item())
    all_scores["langs"].append(lang)

# Separate scores by language
en_scores = {k: [] for k in ["bleu", "rouge1", "rougeL", "meteor", "bertscore_f1"]}
es_scores = {k: [] for k in ["bleu", "rouge1", "rougeL", "meteor", "bertscore_f1"]}

for i, r in enumerate(generated_results):
    lang = r["metrics"]["language"]
    target = en_scores if lang == "en" else es_scores

    for metric in ["bleu", "rouge1", "rougeL", "meteor", "bertscore_f1"]:
        target[metric].append(r["metrics"][metric])

# Function to print averages
def print_avg_scores(title, scores):
    print(f"\n📊 {title} (Avg over {len(scores['bleu'])} samples):")
    for metric in scores:
        avg = sum(scores[metric]) / len(scores[metric]) if scores[metric] else 0
        print(f"{metric}: {avg:.4f}")

# Print results
print_avg_scores("ENGLISH METRICS", en_scores)
print_avg_scores("SPANISH METRICS", es_scores)
```
